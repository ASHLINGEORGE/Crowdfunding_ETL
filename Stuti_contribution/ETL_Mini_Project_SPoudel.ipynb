### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
crowdfunding_info_df.columns
```

```python
# Assign the category and subcategory values to category and subcategory columns.
crowdfunding_info_df[["category", "subcategory"]] = crowdfunding_info_df["category & sub-category"].str.split('/', n =2, expand =True)
crowdfunding_info_df.head()
```

```python
# Get the unique categories and subcategories in separate lists.
categories = crowdfunding_info_df['category'].unique()
print(categories)
subcategories = crowdfunding_info_df['subcategory'].unique()
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 
cat_ids = ["cat"+ str(c) for c in category_ids]
# Use a list comprehension to add "subcat" to each subcategory_id.    
scat_ids = ["subcat"+ str(sc) for sc in subcategory_ids]
    
print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame(
    {"category_id": cat_ids,
     "category":categories
    })

# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_df = pd.DataFrame(
    {"subcategory_id": scat_ids,
     "subcategory":subcategories
    })
```

```python
# Export categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)

subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

### Campaign DataFrame

### Option 1: Use Pandas to create the contacts DataFrame.

```python
#dropping the first row in the contact_info_df
contact_info_df = contact_info_original_df.drop(0)
```

```python
contact_info_df.head(5)
```

```python
# # Get each row into a list and print first 5 rows
contact_info_df_values = contact_info_df.values.tolist()

#print first 5 rows
contact_info_df_values[0:5]
```

```python
# Iterate through the contact_info_df and convert each row to a dictionary.
dict_values = []
# Iterate through each dictionary
for row in contact_info_df_values:
    # Parse the JSON string in each row into a dictionary
    data_dict = json.loads(row[0])    
    # Now, extract the values from the dictionary and append them to new list 'dict_values'
    values = list(data_dict.values())
    dict_values.append(values)
            
#Print out the list of values for each row.
print(dict_values)
```

```python
# Convert the list of dictionaries to JSON format
json_data = json.dumps(dict_values, indent=4)  # indent for pretty printing

# Print or save the JSON data
print(json_data)
```

```python
#Now Get columns into the list

column_names = []

# Assuming the JSON data is in the first column (index 0)
for row in contact_info_df_values:
    # Parse the JSON string in each row into a dictionary
    column_dict = json.loads(row[0])
    
    # Extract the keys (column names) from the dictionary
    keys = list(column_dict.keys())
    
    # Append the keys to the column_names list
    column_names.append(keys)

# Print out the list of column names
print(column_names[0])
```

```python
#Getting a single list of column_names
for column_name in column_names:
    column_name = column_names[0]
print(column_name)
```

```python
# Create a contact_info DataFrame and add each list of values, i.e., each row 
# to the 'contact_id', 'name', 'email' columns.

contacts_info_df = pd.DataFrame(dict_values, columns = column_name)
contacts_info_df.head(5)
```

```python
# Check the datatypes.
contacts_info_df.info()
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
contacts_info_df[["first_name","last_name"]] = contacts_info_df["name"].str.split(" ", n = 4, expand =True)
contacts_info_df.head(10)
# Drop the contact_name column
column_to_drop1= ["name"]
contacts_deleted_df = contacts_info_df.drop(column_to_drop1, axis =1)
contacts_deleted_df
contacts_deleted_df.head(10)
```

```python
# Reorder the columns
contacts_df_clean = contacts_deleted_df[["contact_id", "first_name", "last_name", "email"]]
contacts_df_clean.head(10)
```

```python
# Check the datatypes one more time before exporting as CSV file.
contacts_df_clean.info()
```

```python
# Export the DataFrame as a CSV file. 
contacts_df_clean.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```

### Option 2: Use regex to create the contacts DataFrame.

```python
contact_info_df_copy = contact_info_original_df.copy()
contact_info_df_copy.head()
```

```python
# Set the first row as column names
contact_info_df_copy.columns = contact_info_df_copy.iloc[0]
```

```python
# Drop the first row after setting it as column names
contact_info_df_copy = contact_info_df_copy.iloc[1:]
contact_info_df_copy
# Print the updated DataFrame
contact_info_df_copy[0:5]
```

```python
# Get each row into a list.
contact_info_df_list = contact_info_df_copy.values.tolist()
contact_info_df_list
#print first 5 rows
contact_info_df_list[0:5]
```

```python
# Split the values of the first row in the list on the non-word character to find the 'contact_id' for first row
first_row = re.findall('[0-9]{4}',contact_info_df_list[0][0])
first_row
```

```python
# Extract the four-digit contact ID number.
pattern = r'"contact_id": (\d+)'

# Use list comprehension to extract 'contact_id' from each row
contact_ids = [re.search(pattern, row[0]).group(1) for row in contact_info_df_list]

# Print 'contact_id' values for all rows
print(contact_ids[0:10])
```

```python
# Add a new "contact_id" column to contact_info_df_copy with values from contact_ids list
contact_info_df_copy["contact_id"] = contact_ids

# Print the updated DataFrame
contact_info_df_copy.head(5)
```

```python
# Check the datatypes.
contact_info_df_copy.info()
```

```python
# Convert the "contact_id" column to an int64 data type.
contact_info_df_copy["contact_id"] = contact_info_df_copy["contact_id"].astype('int')
contact_info_df_copy.info()
```

```python
# Extract the name of the contact and add it to a new column.
pattern = r'"name": "([^"]+)"'
# Use list comprehension to extract 'name' from each row
names = [re.search(pattern, row[0]).group(1) if re.search(pattern, row[0]) else None for row in contact_info_df_list]

# Print 'name' values for the first 10 rows
print(names[0:10])
```

```python
# Add a new "name" column to contact_info_df_copy with values from 'names' list
contact_info_df_copy["name"] = names

# Print the updated DataFrame
contact_info_df_copy.head(10)
```

```python
# Extract the email from the contacts and add the values to a new column.
# Extract the emails of the contact and add it to a new column.
pattern = r'"email": "([^"]+)"'

# Use list comprehension to extract 'emails' from each row
emails = [re.search(pattern, row[0]).group(1) if re.search(pattern, row[0]) else None for row in contact_info_df_list]

# Print 'emails' values for the first 10 rows
print(emails[0:10])
```

```python
# Add a new "email" column to contact_info_df_copy with values from 'emails' list
contact_info_df_copy["email"] = emails

# Print the updated DataFrame
contact_info_df_copy.head(10)
```

```python
# Create a copy of the contact_info_df with the 'contact_id', 'name', 'email' columns.
contact_info_copied_df = contact_info_df_copy[['contact_id', 'name', 'email']].copy()

# Print the copied DataFrame
contact_info_copied_df.head(10)
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
pattern = r'(\w+)\s+(\w+)'
contact_info_copied_df[["first_name", "last_name"]] = contact_info_copied_df["name"].str.extract(pattern)
contact_info_copied_df

# Drop the contact_name column
contact_info_copied_df.drop(columns=['name'],inplace=True)
contact_info_copied_df.head(10)
```

```python
# Reorder the columns
contacts_df_clean_regex = contact_info_copied_df[["contact_id", "first_name", "last_name", "email"]]
contacts_df_clean_regex.head(10)
```

```python
# Check the datatypes one more time before exporting as CSV file.
contacts_df_clean_regex.info()
```

```python
# Export the DataFrame as a CSV file. 
contacts_df_clean_regex.to_csv("Resources/contactsregex.csv", encoding='utf8', index=False)
```
